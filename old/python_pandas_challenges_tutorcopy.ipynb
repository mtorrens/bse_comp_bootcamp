## Functions

```python
# Challenge 1: Build a function that receives as input one tuple from our list of tuples (tuples section) , 
# and returns the name and (his age + 10)
def check_age(onetuple):
    return onetuple[0],onetuple[1]+10

# Test it with this function
onetuple=('bob',23)#one of the valid names
check_age(onetuple)
```

```python
# Challenge 2: Apply previous function to all tuples in our list of tuples
output=[]
for tuples in mytuple:
    output.append(check_age(tuples))
print(output)

#or [check_age(onetuple) for onetuple in mytuple]
```

```python
# Challenge 3: Build a function that receives as input one name the dictionary that we built in challenge #3 
# (section dictionaries) and yields a tuple of 'name' and maximum age of best friends
# Test it with some names of our list of dictionaries
def check_best(name, mydict):
    max_best=None
    for one_dict in mydict:        
        if one_dict['name']==name:
            max_best=max(one_dict['best_friends'])
    return max_best

check_best('bob', mydict)
```

```python
# Challenge 4: Modify previus function, so that handles names that are not present
# i.e. when checking a new names, yield a message such as ('Robert is not in our list')
def check_best(name, mydict):
    max_best=None
    for one_dict in mydict:        
        if one_dict['name']==name:
            max_best=max(one_dict['best_friends'])
    if max_best is None:
        max_best=name+ ' is not in our list'
    return max_best


# Test it with a new name
check_best('Joan', mydict)
```

```python
# Challenge 5: Aply previous function to all names in our list of dictionaries
out=[]
for one_dict in mydict:
    myname= one_dict['name']
    mybest=check_best(myname,mydict)
    out.append(mybest)
out

```
