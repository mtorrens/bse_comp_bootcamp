## Exercise 2 [30%]: *A 'catastrophic' draw*

**Background**

Last Dec. 13, soccer fans around the globe were waiting in front of the TV to watch the draw to the final knockout stage of the most important club competition in the world, the UEFA Champions League.

The round of 16 was supposed to be a fairly simple draw: the group stage had eight groups of teams [A–H], and the teams finishing on the top two positions of each group made it into the draw. So, there were two pots of 8 teams to be paired, in order to determine which teams would face each other to kick off the final stage of the competition. In the draw, only three simple conditions had to be met:

*   Each group champion had to be drawn against a group runner-up
*   Two teams from the same group could not be drawn together
*   Two teams coming from the same football federation could not be drawn together

Surprisingly enough, the "computer" made a mistake and [the draw was a disaster](https://www.goal.com/en/news/champions-league-draw-what-happened-why-uefa-redo/bltdeae17cc8a4b1c83), so much that it was declared void and had to be repeated. Given the magnitude and simplicity of the draw, this was nothing short of unbelievable.




**The task**

We would like to know what were the probabilities of teams facing each other prior to the actual draw. Since this is not a probability course, the way we choose to proceed is to use simulation: we will replicate the original draw for a large number of times and estimate the probabilities of the different matchups with their long run frequency as given by the proportions of times they occurred in our "large" sample.

Simulate the draw $R = 10^{6}$ times, i.e. produce one million realisations of the original draw setup. Once you have your draw sample, proceed to **Q22** to **Q27**.
