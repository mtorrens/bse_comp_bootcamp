## 2. Google Colab

We will be working in Google Colab throughout the course. You can find more details about Colab [here](https://colab.research.google.com/notebooks/basic_features_overview.ipynb).

To open a new notebook in Google Colab simply go to `New` > `More` > `Google Colaboratory`, and you are good to go. You can add chunks of text and code using simple tabs. New notebooks are by default loaded with a `python` engine. If you want them to run with `R`, go to `Edit` > `Notebook settings` > `Runtime type`, and select `R`.

You can also save and download notebooks in `.ipynb` format, share on Github, etc. Plus, everything is fairly customisable. If you're new to Colab, I suggest you take a few minutes to familiarise with its contents.
